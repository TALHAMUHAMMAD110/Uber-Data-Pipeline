```python
import pandas as pd
import datetime as dt
df = pd.read_parquet('data/yellow_tripdata_2025-01.parquet')
df.head(10)
```

```python
datetime_dim_table = df[['tpep_dropoff_datetime','tpep_pickup_datetime']].drop_duplicates().reset_index(drop = True)
```

```python
datetime_dim_table['pick_hour'] = datetime_dim_table['tpep_pickup_datetime'].dt.hour
datetime_dim_table['pick_month'] = datetime_dim_table['tpep_pickup_datetime'].dt.month
datetime_dim_table['pick_year'] = datetime_dim_table['tpep_pickup_datetime'].dt.year
datetime_dim_table['pick_weekday'] = datetime_dim_table['tpep_pickup_datetime'].dt.weekday


datetime_dim_table['drop_hour'] = datetime_dim_table['tpep_dropoff_datetime'].dt.hour
datetime_dim_table['drop_month'] = datetime_dim_table['tpep_dropoff_datetime'].dt.month
datetime_dim_table['drop_year'] = datetime_dim_table['tpep_dropoff_datetime'].dt.year
datetime_dim_table['drop_weekday'] = datetime_dim_table['tpep_dropoff_datetime'].dt.weekday

datetime_dim_table
```
